```python
def plot_hist(title,series,color,edge_min,edge_max,intv,ax):
    bins = (edge_max-edge_min)//intv
    ax.grid(zorder=0)
    if color is not None:
        ax.hist(series, bins=bins,color=color,range=(edge_min,edge_max),zorder=1)
    else:
        ax.hist(series, bins=bins,range=(edge_min,edge_max),zorder=1)
    ax.set_title(title)
def filename_mapping(rule_name,dataset_name,pos_num,neg_num):
    return '%s/%s_%d_%d.csv'%(rule_name,dataset_name,pos_num,neg_num)
def generate_datasets(train_num_tls,test_num_tls):
    rules = ['rule2']
    for rule_name in rules:
        for pos_num,neg_num in train_num_tls:
            filename = filename_mapping(rule_name,'train',pos_num,neg_num)
            generate_dataset(pos_num,neg_num,rule_name,filename)
        for pos_num,neg_num in test_num_tls:
            filename = filename_mapping(rule_name,'test',pos_num,neg_num)
            generate_dataset(pos_num,neg_num,rule_name,filename)
def read_datasets(rule_name,dataset_name,pos_neg_list):
    df_list = []
    for pos_num,neg_num in pos_neg_list:
        path = filename_mapping(RULE_NAME,dataset_name,pos_num,neg_num)
        df_list.append(read_dataset(path))
    return df_list
def visualize_tree(clf,feature_names=["volume","mass","length","shape","color","rough"]):
    from sklearn.externals.six import StringIO  
    import pydotplus 
    from sklearn import tree
    dot_data = tree.export_graphviz(clf, out_file= None, 
                         feature_names=feature_names,  
                         class_names=["dark","light"],  
                         filled=True, rounded=True)  
    graph = graphviz.Source(dot_data)
    display(graph)
```

```python
def plot_target_variable(df_list):
    num_df = len(df_list)
    fig, axs  = plt.subplots(num_df, 2,figsize=(10,5))
    for i,df in enumerate(df_list):
        df['light'].value_counts().plot.pie(autopct='%1.1f%%',shadow=True,ax=axs[i][0])
        sns.countplot(x='light',data=df,ax=axs[i][1])
        axs[i][0].set_title('dataset %d'%(i+1))
        axs[i][1].set_title('dataset %d'%(i+1))
    plt.tight_layout()
plot_target_variable(train_df_list)
```

```python
def plot_value_feature(df_list,titles,colors,min_val,max_val,intv,y_ticks,y_lim):
    num_df = len(df_list)
    fig, axs  = plt.subplots(1,num_df,figsize=(10,5))
    for i,df in enumerate(df_list):
        ax = axs[i]
        if y_ticks is not None:
            ax.set_yticks(y_ticks)
        if y_lim is not None:
            ax.set_ylim(y_lim)   
        plot_hist(titles[i],df,colors[i],min_val,max_val,intv,ax)
    plt.tight_layout()

def plot_cat_feature(df_list,feature,titles):
    num_df = len(df_list)
    fig, axs  = plt.subplots(1,num_df,figsize=(10,5))
    for i,df in enumerate(df_list):
        sns.countplot(x="light",hue=feature, data=df,ax=axs[i])
        axs[i].set_title(titles[i])
    plt.tight_layout()
```

## Decision tree
第一張圖為用100筆正樣本和100筆負樣本訓練產生的決策樹  
第二張圖為用100筆正樣本和30筆負樣本訓練產生的決策樹

第一個Decision tree的feature importance為  
volume=0.7      mass=0.1   length=0.03    shape=0     color=0      rough=0.1] 
而判斷的方式以volume,mass,length,rough的線性組合的值決定類別,符合規則

而第二個Decision tree的feature importance在shape和color的部分皆有值,然而規則無關,
可能是因為正負樣本不均勻或資料集比較小導致其overfitting的關係

而每一張圖下面有兩行accuracy代表這顆決策樹(模型)在測試資料集1(正樣本100,負樣本100)和測試資料集2(正樣本100,負樣本30)的表現(以下的其他模型也是)

```python
train_df_list = read_datasets(RULE_NAME,'train',[(100,100),(100,30)])
test_df_list =  read_datasets(RULE_NAME,'test',[(100,100),(100,30)])
def feature_engineering(df):
    X,y = df.iloc[:, :-1].values,df.iloc[:, -1].values
    return X,y

print('decision tree')
for train_df in train_df_list:
    train_X,train_y = feature_engineering(train_df)
    clf =  DecisionTreeClassifier()
    clf.fit(train_X,train_y)
    print(clf.feature_importances_)
    visualize_tree(clf)
    for test_df in test_df_list:
        test_X,test_y = feature_engineering(test_df)
        evaluate(test_X,test_y,clf)
```

# 加入特徵
接下來的實驗觀察加入一個和規則產生有高度相關的特徵後  
mass/volume (density) 觀察模型的表現會有多大程度的提升

```python
def feature_engineering_magic(df):
    series = df['mass']+df['volume']-df['length']
    if 'magic' not in df.columns:
        df.insert(2, 'magic', series)
    X,y = df.iloc[:, :-1].values,df.iloc[:, -1].values
    return X,y
```

# Decision tree
除了 performance提升外,規則也變簡單了,因為可以直接依靠magic特徵

```python
print('decision tree')
for train_df in train_df_list:
    train_X,train_y = feature_engineering_magic(train_df)
    clf =  DecisionTreeClassifier()
    clf.fit(train_X,train_y)
    print(clf.feature_importances_)
    visualize_tree(clf,["volume","mass","magic","length","shape","color","rough"])
    for test_df in test_df_list:
        test_X,test_y = feature_engineering_magic(test_df)
        evaluate(test_X,test_y,clf)
```

# Random forest
除了 performance提升外,Feature importance的分布更符合產生資料的規則

```python
print('random forest')
for train_df in train_df_list:
    train_X,train_y = feature_engineering_magic(train_df)
    clf = RandomForestClassifier(n_estimators=5000)
    clf.fit(train_X,train_y)
    print(clf.feature_importances_)
    for test_df in test_df_list:
        test_X,test_y = feature_engineering_magic(test_df)
        evaluate(test_X,test_y,clf)
```

# Logistic regression
雖然performance提升但是有在不應該給負的權重的feture上給上負的權重,是因為overfitting的原因嗎(特徵太多)

```python
print('logistic regression')
for train_df in train_df_list:
    train_X,train_y = feature_engineering_magic(train_df)
    #clf = LogisticRegression(penalty='l1')
    clf = LogisticRegression(penalty='l2')
    clf.fit(train_X,train_y)
    print(clf.coef_)
    for test_df in test_df_list:
        test_X,test_y = feature_engineering_magic(test_df)
        evaluate(test_X,test_y,clf)
```
